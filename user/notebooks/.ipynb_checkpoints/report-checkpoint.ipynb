### Support Functions

```python
#@title
def get_sparql_dataframe(service, query):
    """
    Helper function to convert SPARQL results into a Pandas data frame.
    """
    sparql = SPARQLWrapper(service)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    result = sparql.query()

    processed_results = json.load(result.response)
    cols = processed_results['head']['vars']

    out = []
    for row in processed_results['results']['bindings']:
        item = []
        for c in cols:
            item.append(row.get(c, {}).get('value'))
        out.append(item)

    return pd.DataFrame(out, columns=cols)
```

## Queries

What follows is a set of queries designed to provide a feel for the OIH graph

### Simple Triple Counts


```python
cq = """
SELECT (COUNT(*) as ?Count) 
WHERE 
  {
      { ?s ?p ?o } 
  }
"""

dfsc = get_sparql_dataframe(iow, cq)
dfsc.head()
```

### Type Count

How many triples are there?

```python
rq_count = """select  ?o (COUNT(?o) as ?oCount)
where { 
	?s a ?o .
} 
GROUP By ?o
ORDER BY DESC(?oCount)
"""
```

```python
dfsc = get_sparql_dataframe(iow, rq_count)
dfsc.set_index('o', inplace=True)
dfsc.head(20)
```

### Predicate Counts

This gives an overview of unique predicates that connect a subject to an object.  This gives us both an idea of the properties we are using on things and count of their usage.


```python
rq_pcount = """SELECT ?p (COUNT(?p) as ?pCount)
WHERE
{
  ?s ?p ?o .
}
GROUP BY ?p
ORDER BY DESC(?pCount)
"""
```

```python
dfc = get_sparql_dataframe(iow, rq_pcount)
dfc['pCount'] = dfc["pCount"].astype(int) # convert count to int
dfc.set_index('p', inplace=True)
dfc.head(20)
```

```python
hs = dfc.head(10)
plot = hs.plot.pie(y='pCount',x='p',legend=False, figsize=(10, 10))
```

## Try geosparql

```python
rq_main = """PREFIX my: <http://example.org/ApplicationSchema#>
PREFIX geo: <http://www.opengis.net/ont/geosparql#>
PREFIX geof: <http://www.opengis.net/def/function/geosparql/>

SELECT *
WHERE {
    ?s ?p ?fGeom .
    ?fGeom geo:asWKT ?fWKT .
    FILTER (geof:sfWithin(?fWKT, '''
        <http://www.opengis.net/def/crs/OGC/1.3/CRS84>
           POLYGON((-129.48921751581898 51.721912862764334,-59.17671751581898 51.721912862764334,-59.17671751581898 21.445382674907275,-129.48921751581898 21.445382674907275,-129.48921751581898 51.721912862764334))
        '''^^geo:wktLiteral))
} 
        """
```

```python
df = get_sparql_dataframe(iow, rq_main)
df['fWKT'] = df['fWKT'].apply(wkt.loads)
# df.head(5)
```

```python
gdf = geopandas.GeoDataFrame(df, geometry='fWKT', crs={"init": "epsg:3857"}) # contextily requires 3875?
```
